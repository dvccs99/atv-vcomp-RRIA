### Segmentação

Selecione uma rede pré-treinada de segmentação e avalie os resultados para 2, 3, 4 e 5 classes em imagens coloridas. Utilize no mínimo 5 imagens.

```python
import segmentation_models_pytorch as sm
import matplotlib.pyplot as plt
import cv2
import numpy as np
from PIL import Image
```

```python
model = sm.Unet()
```

```python
# Configurar layout (1 linha x 5 colunas)
rows, cols = 1, 5

# Criar a figura
fig, axes = plt.subplots(rows, cols, figsize=(15, 5))

# Carregar as imagens
circuito = plt.imread("dvccs_imgs/circuito.jpg")
beagle = plt.imread("dvccs_imgs/beagle.jpg")
denso = plt.imread("dvccs_imgs/denso.jpg")
flores = plt.imread("dvccs_imgs/flores.jpg")
gatos = plt.imread("dvccs_imgs/gatos.jpg")

# Lista de imagens
imgs = [circuito, beagle, denso, flores, gatos]

# Rotacionar e exibir as imagens
for i, img in enumerate(imgs):
    # Rotacionar a imagem (se necessário)
    if i != 4:
        img = np.rot90(img, k=3)
    
    # Exibir a imagem no subplot correspondente
    axes[i].imshow(img)
    axes[i].axis("off")  # Remover os eixos

# Ajustar espaçamento
plt.tight_layout()
plt.show()
```
